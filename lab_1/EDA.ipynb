```python
from itertools import combinations
from collections import Counter
import networkx as nx
from pyvis.network import Network
from IPython.display import IFrame

with open("lab_task_repo/BDSnowflake/исходные данные/MOCK_DATA.csv") as f:
    columns = f.readline().strip().split(",")

stopwords = {"id", "name", "date"}
token_freq = Counter()
pairs = Counter()

for col in columns:
    parts = [p for p in col.lower().split("_") if p not in stopwords]
    for p in parts:
        token_freq[p] += 1
    for a, b in combinations(sorted(set(parts)), 2):
        pairs[(a, b)] += 1

G = nx.Graph()
for token, freq in token_freq.items():
    G.add_node(token, size=freq)

for (a, b), w in pairs.items():
    G.add_edge(a, b, weight=w)

net = Network(
    height="950px",
    width="100%",
    bgcolor="white",
    font_color="black",
    notebook=False,
    cdn_resources="in_line"
)

net.set_options("""
var options = {
  "nodes": {
    "shape": "dot",
    "size": 18,
    "font": {
      "size": 34,
      "face": "arial",
      "color": "#111111",
      "strokeWidth": 10,
      "strokeColor": "#ffffff"
    },
    "scaling": {
      "label": {
        "enabled": true,
        "min": 26,
        "max": 42,
        "maxVisible": 42,
        "drawThreshold": 1
      }
    }
  },
  "edges": {
    "width": 1.0,
    "smooth": false,
    "color": {
      "color": "#9aa4b2",
      "highlight": "#4f46e5"
    }
  },
  "interaction": {
    "hover": true,
    "tooltipDelay": 100
  },
  "physics": {
    "barnesHut": {
      "gravitationalConstant": -8000,
      "centralGravity": 0.12,
      "springLength": 180,
      "springStrength": 0.015,
      "damping": 0.2
    },
    "minVelocity": 0.75
  }
}
""")

for node, data in G.nodes(data=True):
    net.add_node(
        node,
        label=node,
        title=f"{node}: freq={data['size']}",
        size=12 + data["size"] * 3,
        font={
            "size": 34,
            "color": "#111111",
            "strokeWidth": 10,
            "strokeColor": "#ffffff",
            "face": "arial",
        }
    )

for u, v, data in G.edges(data=True):
    net.add_edge(
        u, v,
        value=data["weight"],
        title=f"{u} ↔ {v}: {data['weight']}"
    )

net.write_html("columns_graph.html")
```
